# Exploratory Data Analysis

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import random
```

```python
# define paths
path_train_img = '../data/processed_kolektor/train/images'
path_test_img = '../data/processed_kolektor/test/images'
path_mask_img = '../data/processed_kolektor/test/masks'
```

```python
# count samples
print(f'Training samples: {len(os.listdir(path_train_img))}')
print(f'Testing samples: {len(os.listdir(path_test_img))}')
```

```python
# pick test samples and shuffle them
samples = os.listdir(path_test_img)
random.shuffle(samples)
```

```python
# show some samples
shown = 0
show_total = 5
for fname in samples:
    base = fname[:-4]
    mask_path = os.path.join(path_mask_img, base + '_label.bmp')
    mask = np.array(Image.open(mask_path).convert('L'))
    if mask.max() == 0:
        mask_title = 'Mask with no defect'
    else:
        mask_title = 'Mask with defect'

    img = np.array(Image.open(os.path.join(path_test_img, fname)))
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.title('Image')
    plt.imshow(img)
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.title(mask_title)
    plt.imshow(mask, cmap='gray')
    plt.axis('off')
    plt.show()
    shown += 1
    if shown >= show_total:
        break
```
